```
import pandas as pd
import matplotlib.pyplot as plt

survey = pd.read_csv("../data/survey_data.csv")
chat = pd.read_csv("../data/chat_feedback.csv")
nps = pd.read_csv("../data/nps_scores.csv")

# Merge survey + nps
merged = survey.merge(nps, on="customer_id", suffixes=("_survey", "_nps"))
print(merged.head())

# Rating distribution
plt.hist(merged['rating'], bins=5)
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

# NPS distribution
plt.hist(merged['nps_score'], bins=5)
plt.title('NPS Score Distribution')
plt.xlabel('NPS Score')
plt.ylabel('Count')
plt.show()
```
